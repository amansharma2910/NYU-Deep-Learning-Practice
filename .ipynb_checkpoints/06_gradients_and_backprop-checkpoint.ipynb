# Backpropagation and Autograd
---

In this notebook, we are going to take a look at how backpropagation works in PyTorch, along with the autograd module of PyTorch that performs automated gradient calculation on a given tensor operation. 
Autograd is a define-by-run operation, which simply means that the gradients are calculated on the basis of the order of calculations that you perform in your model function.

Let us begin this tutorial by importing all the necessary modules.

```python
import torch
from torchviz import make_dot
```

Now, let us define a tensor then some linear transformation on the defined tensor.

```python
# defining the tensor
x = torch.tensor([[1, 2],[3, 4]], requires_grad=True, dtype=torch.float32)
x
```

```python
# defining linear transformation
y = x + 2
```

Since y is a dependent function defined on the independent variable x, therefore y has a gradient function. To check this gradient function, we will use the `grad_fn` attribute.

```python
y.grad_fn
```

```python
# since x is an independent variable, it's grad function is none 
x.grad_fn
```

```python
y.grad_fn.next_functions[0][0].variable
```

```python
make_dot(y)
```

Now, let us try to perform a similar operation but on a different, relatively more complex operation.

```python
z = y**2 + 3
a = z.mean()

print(x)
print(a)
```

```python
make_dot(a)
```

Now let us perform autograd operation on a. This will simply calculate the gradients of the a (our dependent variable) w.r.t. x (dependent variable).

For calculating gradients in PyTorch, we use the `backward()` method on the vector/matrix we want to calculate the gradients for w.r.t. the parameters it is dependent on. 

```python
# computing gradients for a
a.backward()
```

The gradients of a w.r.t. x can be accessed using the `grad` attribute of the matrix x.

```python
# accessing gradients of a w.r.t. x
x.grad
```

```python
x
```

One thing to be noted here is that the gradients get accumulated only for those tensors for which we have set the `requires_grad` parameter as `True`. 

Let us see this with the help of an example.

```python
# Only w that allows gradient accumulation
x = torch.arange(1., 4)
w = torch.ones(3, requires_grad=True)
z = w @ x
z.backward()
print(x.grad, w.grad, sep='\n')
```

One thing to be noted here is that the autograd in PyTorch tracks (and generates computational graphs for) all the mathematical operations by default. However, you can turn off this default be behavior using the `torch.no_grad()`. It is especially a good practive to turn off computational graph generation during the validation/inference phase, since the inference phase does not require gradient calculation or backpropagation for parameter updation.

However, you need to keep in mind that running the backward() operation on a matrix/vector for which gradient calculation is turned off will result in a runtime error.

Let us see this with the help of an example.

```python
m = torch.arange(1, 4, dtype=torch.float32)
n = torch.ones(3, requires_grad=True)
with torch.no_grad():
    k = n @ m

# performing backprop (will result in an error)
try:
    k.backward()
except RuntimeError as e:
    print(e)
```

With this, we come to an end of this tutorial where we got a hands-on on how autograd and gradient calculations work in PyTorch.
